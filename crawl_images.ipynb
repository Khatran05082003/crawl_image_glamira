```python
import requests
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
import random


session = requests.Session()
retry_strategy = requests.packages.urllib3.util.retry.Retry(
    total=3,  
    status_forcelist=[429, 500, 502, 503, 504],  
    allowed_methods=["HEAD", "GET", "OPTIONS"],  
    backoff_factor=random.uniform(0, 10)  
)
adapter = requests.adapters.HTTPAdapter(max_retries=retry_strategy)
session.mount("https://", adapter)
session.mount("http://", adapter)

base_url = 'http://cmedia/sitemap/glus/product_provider-41-{}.xml'  

sitemap_numbers = range(1, 10)

all_urls = []

for num in sitemap_numbers:
    xml_url = base_url.format(num)
    try:
        response = session.get(xml_url)
        response.raise_for_status()  
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch {xml_url}: {e}")
        continue
    
    xml_content = response.content
    root = ET.fromstring(xml_content)
    
    ns = {'ns': 'http://www.sitemaps.org/schemas/sitemap/0.9'}
    urls = [url.find('ns:loc', ns).text for url in root.findall('ns:url', ns)]
    all_urls.extend(urls)

print(len(all_urls))

def fetch_image_sources(url):
    try:
        response = session.get(url)
        response.raise_for_status()  
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch {url}: {e}")
        return []
    
    soup = BeautifulSoup(response.content, 'html.parser')
    img_tags = soup.find_all('img')
    img_sources = [img['src'] for img in img_tags if 'src' in img.attrs]
    
    return img_sources

all_image_sources = []
for url in all_urls:
    image_sources = fetch_image_sources(url)
    all_image_sources.extend(image_sources)
    print(f"Found {len(image_sources)} images at {url}")

print(len(all_image_sources))
```
